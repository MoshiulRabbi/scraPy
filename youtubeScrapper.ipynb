```python
from itertools import islice
from youtube_comment_downloader import *
import requests
import re
import json
import numpy as np
import pandas as pd
```

```python
#@title Youtube Scrapper

youtubeLink = 'https://www.youtube.com/watch?v=MhCxKETP_Z8' #@param {type:"string"}
numberOfComments = 100 #@param {type: "integer"}

c = []
cf = []
downloader = YoutubeCommentDownloader()
comments = downloader.get_comments_from_url(youtubeLink, sort_by=SORT_BY_POPULAR)
for comment in islice(comments, numberOfComments):
    c.append(comment)

for i in c:
  cf.append(i['text'])
```

```python
df = pd.DataFrame(cf, columns=["comments"])
df
```

```python
#only bangla comments

chars_bengali = [chr(c) for c in range(0x0980, 0x09ff)]
df_bengali = df[df.comments.str.contains('|'.join(chars_bengali)).fillna(False)]
df_bengali.to_csv('bn.csv', index=False)
bn = pd.read_csv('bn.csv')
bn
```

```python
resultExcelFile = pd.ExcelWriter('ResultExcelFile.xlsx')

# converting the csv file to an excel file
bn.to_excel(resultExcelFile, index=False)

# saving the excel file
resultExcelFile.save()
```
